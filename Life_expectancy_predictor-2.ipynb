```python
#Dependencies

import pandas as pd
from matplotlib import pyplot
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
```

```python
# Load file to data frame

df_life_expectancy = pd.read_csv("Resources/cleaned_file.csv")
df_life_expectancy.head(10)
```

```python
# Set the combination of Country + Year as the Index for the data frame

df_life_expectancy.set_index('Country_Year', inplace=True)
df_life_expectancy.head(10)
```

```python
#Check column names

df_life_expectancy.columns
```

```python
# Remove extra spaces in column names

df_life_expectancy.rename(columns={'Life expectancy ': 'Life expectancy'}, inplace=True)
df_life_expectancy.rename(columns={'Measles ': 'Measles'}, inplace=True)
df_life_expectancy.rename(columns={'BMI ': 'BMI'}, inplace=True)
df_life_expectancy.rename(columns={'under-five deaths ': 'Under-five deaths'}, inplace=True)
df_life_expectancy.rename(columns={'Diphtheria ': 'Diphtheria'}, inplace=True)
df_life_expectancy.rename(columns={' thinness  1-19 years': 'Thinness 1-19 years'}, inplace=True)
df_life_expectancy.rename(columns={' thinness  5-9 years': 'Thinness 5-9 years'}, inplace=True)
```

```python
df_numeric = df_life_expectancy.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = df_numeric.corr()

# Get the correlation of each variable with Life Expectancy
life_expectancy_correlation = correlation_matrix['Life expectancy'].sort_values(ascending=False)

# Display the results
print("Correlation with Life Expectancy:")
print(life_expectancy_correlation)
```

```python
# Select only the numeric columns 
df_numeric = df_life_expectancy.select_dtypes(include=['float64', 'int64'])

# Define the target variable 
y = df_numeric['Life expectancy']


X = df_numeric.drop(columns=['Life expectancy'])

# Initialize the linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Get the coefficients 
coefficients = pd.DataFrame(model.coef_, X.columns, columns=['Coefficient'])
coefficients = coefficients.round(3)

# Display the coefficients
print("\nLinear Regression Coefficients:")
print(coefficients.sort_values(by='Coefficient', ascending=False))
```

```python
# Check data types

df_life_expectancy.dtypes
```

```python
# Set the Status column data type to Category, for use with XGBoost's experimental categorical variable functionality

df_life_expectancy['Status'] = df_life_expectancy['Status'].astype('category')
```

```python
# Check for null values in the Life expectancy column

df_life_expectancy['Life expectancy'].isnull().sum()
```

```python
# Drop all rows where Life expectancy is null

df_life_expectancy = df_life_expectancy.dropna(subset=['Life expectancy'])

# Check for null values in the Life expectancy column

df_life_expectancy['Life expectancy'].isnull().sum()
```

```python
# Define input and target columns

y = df_life_expectancy['Life expectancy']
x = df_life_expectancy.drop(columns = ['Life expectancy', 'Country', 'Year'])
```

```python
# Generate training and test splits
# Note: The stratify paramter of test_train_split cannot be used because of NaN values in data set

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1, test_size=0.20)
```

```python
# Initialize XGBoost Regressor model

model = XGBRegressor(enable_categorical=True)
```

```python
# Fit model to training split

model.fit(x_train, y_train)
```

```python
# Make predictions

y_predict = model.predict(x_test)
```

```python
# Evaluate model accuracy

mse = mean_squared_error(y_test, y_predict)
r_squared = r2_score(y_test, y_predict)
mape = mean_absolute_percentage_error(y_test, y_predict)

# Print model metrics

print(f"The R-squared value is: {round(r_squared, 3)}")
print(f"The mean squared error is: {round(mse, 3)}")
print(f"The mean absolute percentage error is: {round(mape, 3)}")
```
